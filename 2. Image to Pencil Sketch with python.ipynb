```python
import numpy as np
import pandas as pd
import cv2 as cv
from PIL import Image
import matplotlib.pyplot as plt
from IPython.display import display
```

# Original Images

```python
img = cv.imread('tiger.jpg')
display(Image.fromarray(img))                 
```

# Converting the Original image into Grayscale

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
display(Image.fromarray(gray))
```

# Inverted Grayscale image as negative image

```python
inv_gray_image=255-gray
display(Image.fromarray(inv_gray_image))
```

# Blurred Image

```python
blur=cv.GaussianBlur(inv_gray_image,(35,35), 0)
display(Image.fromarray(blur))
```

#  Inverted Blurred Image

```python
inverted_blur_image=255-blur
display(Image.fromarray(inverted_blur_image))
```

# Creating the pencil Sketch Image

```python
pencil_sketch=cv.divide(gray,inverted_blur_image,scale=256)
```

```python
cv.imwrite("output.png",pencil_sketch)
display(Image.fromarray(pencil_sketch))
```
